```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
coefficient = model.coef_[0]
intercept = model.intercept_
print(f'Coefficient: {coefficient}')
print(f'Intercept: {intercept}')
```

```python
# Training set
train_predictions = model.predict(X_train)
mse_train = mean_squared_error(y_train, train_predictions)
r2_train = r2_score(y_train, train_predictions)
```

```python
# Testing set
test_predictions = model.predict(X_test)
mse_test = mean_squared_error(y_test, test_predictions)
r2_test = r2_score(y_test, test_predictions)
```

```python
print(f'Mean Squared Error (Training): {mse_train}')
print(f'R-squared (Training): {r2_train}')
print(f'Mean Squared Error (Testing): {mse_test}')
print(f'R-squared (Testing): {r2_test}')
```


The linear regression model is used to predict the median value of owner-occupied homes (MValue) based on the average number of rooms (Avg_rooms) demonstrates a positive relationship between these variables. The model suggests that, on average, each additional room is associated with an increase of approximately $8,926 in the median home value. The model's R-squared value of 0.54 indicates that it captures 54% of the variance in MValue using Avg_rooms. The mean squared error on the testing set is 33.85, reflecting the average squared difference between predicted and actual values. While the linear regression provides a useful baseline understanding, it's crucial to acknowledge the simplifications inherent in the model, as housing prices are influenced by a multitude of factors.
